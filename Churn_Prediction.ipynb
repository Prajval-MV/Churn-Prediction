## Now we will begin Model building and deployment process

```python
# Dropping the surname column as there are too many unique values and it is unrelated to churn
data = data.drop(columns=['Surname','id'])
data.head()
```

```python
data_encoded = pd.get_dummies(data) # The get dummies function does label-encoding and one-hot encoding on the categorical data
data_encoded.head()
```

```python
x = data_encoded.drop(columns='Exited')
x
```

```python
y = data['Exited']
y
```

## Testing with Decision Tree Classification Model

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from imblearn.combine import SMOTEENN

# Lets first scale the data by oversampling it since there are way more non-churners than churners

sm = SMOTEENN()
x_rs, y_rs = sm.fit_resample(x,y)
x_train, x_test, y_train, y_test = train_test_split(x_rs,y_rs)

dtc = DecisionTreeClassifier(max_depth=8)
dtc.fit(x_train,y_train)
y_pred = dtc.predict(x_test)
print(classification_report(y_test, y_pred))
```

## Testing with Logistic regression model

```python
from sklearn.linear_model import LogisticRegression
lgm = LogisticRegression()
lgm.fit(x_train,y_train)
y_pred = lgm.predict(x_test)
print(classification_report(y_test,y_pred))
```

### Decision tree seems to be giving a much better result than logistic regression

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
print(classification_report(y_test,y_pred))
```

### We can see KNN is giving a pretty good result with 3 neighbours

## Lets try using some ensemble techniques

```python
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
y_pred=xgb.predict(x_test)
print(classification_report(y_test,y_pred))
```

### Error Analysis of the models

```python
# Lets perform error analysis of all the models that we are testing

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_auc_score 

def ErrorAnalysis(models,num):
    fig, ax = plt.subplots(1,num,figsize=(10, 5))  
    for model in models:
        y_pred = model.predict(x_test)
        m = confusion_matrix(y_pred, y_test, normalize='all')
        disp = ConfusionMatrixDisplay(confusion_matrix=m, display_labels=model.classes_)
        disp.plot(ax=ax[models.index(model)], cmap=plt.cm.Blues, values_format='.4f',colorbar=False)
        ax[models.index(model)].set_title(type(model).__name__)

    plt.tight_layout()
    fig.suptitle('Error Analysis Confusion Matrices',fontsize=14)
    plt.show()

ErrorAnalysis([dtc,lgm,rfc,xgb,knn],5)
```

#### We can see that KNN is better for true positives and XGB is better for true negatives

```python
# Trying to build a soft voting model using the two models with highest accuracy,  XGB and KNN

vc = VotingClassifier([('clf1',knn),('clf2',xgb)],voting='soft')
vc.fit(x_train,y_train)
y_pred = vc.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
ErrorAnalysis([rfc,xgb,knn,vc],4)
```

## We can see that the Voting Classifier is giving the best results !!

```python
#Checking the Test AUC score
vc_preds = vc.predict_proba(x_test)[:,1]                    
print('Test AUC Score',roc_auc_score(y_test,vc_preds))
```
